# Evaluating the ML Models

Now that we finished building the models, lets make predictions and evaluate them:

```python
import joblib
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_pickle('cleaned_apartment_df.pkl')
```

```python
df.head()
```

```python
df = df.drop(columns=['region'])
```

```python
# Define the global models and their corresponding states
global_model_states = {
    'global_model.pkl' : ['AZ', 'CO', 'KS', 'KY', 'LA', 'MD', 'NE', 'NV', 'OH', 'OK', 'TN', 'TX', 'VA', 'WA', 'HI', 'AL'],
    'global_model_2.pkl' : ['FL', 'SC'],
    'global_model_3_1.pkl' : ['NM', 'WV', 'MO', 'NJ', 'IN', 'SD', 'WY', 'MT'],
    'global_model_4.pkl' : ['AK', 'MS', 'VT'], 
    'global_model_5.pkl' : ['DC', 'GA', 'OR'],
    'global_model_6.pkl' : ['ID', 'MA', 'WI'], 
    'combined_states_pipeline.pkl' : ['IL', 'PA', 'RI', 'MI', 'MN', 'NY', 'DE', 'CT']
}
```

```python
# Define the individual model states separately as they do not include the column 'state'
model_states = {
    'AR_model.pkl' : ['AR'],
    'CA_model.pkl' : ['CA'],
    'IA_model.pkl' : ['IA'],
    'NC_model.pkl' : ['NC'],
    'ND_model.pkl' : ['ND'],
    'NH_model.pkl' : ['NH'],
    'UT_model.pkl' : ['UT'],
    'ME_model.pkl' : ['ME']
    }
```

```python
# Function to calculate RMSE
def calculate_rmse(y_true, y_pred):
    return mean_squared_error(y_true, y_pred, squared=False)
```

```python
# Placeholder for results
state_performance = []

# Loop through each model and its corresponding states
for model_filename, states in global_model_states.items():
    # Load the model
    model = joblib.load(model_filename)
    
    for state in states:
        # Filter the dataset for the state
        state_data = df[df['state'] == state]
        X_state = state_data[['bathrooms', 'bedrooms', 'square_feet', 'latitude', 'longitude', 'state']]
        y_state = state_data['price']
        
        # Train-test split for evaluation
        X_train, X_test, y_train, y_test = train_test_split(X_state, y_state, test_size=0.2, random_state=42)
        
        # Make predictions
        y_pred = model.predict(X_test)
        
        # Calculate RMSE
        rmse = calculate_rmse(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        # Store the performance metrics
        state_performance.append({
            'state': state,
            'model': model_filename,
            'RMSE': rmse,
        })
        print(f'State: {state} - Model: {model_filename} - RMSE: {rmse:.2f}, R-squared: {r2:.2f}')
```

```python
# Convert the performance metrics to a DataFrame
global_models_state_performance_df = pd.DataFrame(state_performance)
print(global_models_state_performance_df)
```

```python
# Placeholder for results
individual_state_performance = []

# Loop through each model and its corresponding states
for model_filename, states in model_states.items():
    # Load the model
    model = joblib.load(model_filename)
    
    for state in states:
        # Filter the dataset for the state
        state_data = df[df['state'] == state]
        X_state = state_data[['bathrooms', 'bedrooms', 'square_feet', 'latitude', 'longitude']]
        y_state = state_data['price']
        
        # Train-test split for evaluation
        X_train, X_test, y_train, y_test = train_test_split(X_state, y_state, test_size=0.2, random_state=42)
        
        # Make predictions
        y_pred = model.predict(X_test)
        
        # Calculate RMSE
        rmse = calculate_rmse(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        # Store the performance metrics
        individual_state_performance.append({
            'state': state,
            'model': model_filename,
            'RMSE': rmse,
        })
        print(f'State: {state} - Model: {model_filename} - RMSE: {rmse:.2f}, R-squared: {r2:.2f}')
```

```python
# Convert the performance metrics to a DataFrame
individual_models_state_performance_df = pd.DataFrame(individual_state_performance)
print(individual_models_state_performance_df)
```

```python
models_performance_by_state = pd.concat([global_models_state_performance_df, individual_models_state_performance_df], axis=0)
```

```python
models_performance_by_state = models_performance_by_state.reset_index(drop=True)
```

```python
models_performance_by_state = models_performance_by_state.sort_values(by='state').reset_index(drop=True)
```
